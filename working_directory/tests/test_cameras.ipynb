```python
import cv2

cap = cv2.VideoCapture(1200)
cap2 = cv2.VideoCapture(1201)

while True:
    ret, frame = cap.read()
    ret2, frame2 = cap2.read()
    if not ret or not ret2:
        break

    cv2.imshow("Camera", frame)
    cv2.imshow("Camera 2", frame2)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cap2.release()
cv2.destroyAllWindows()
```
